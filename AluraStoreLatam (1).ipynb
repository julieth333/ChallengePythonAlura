### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print("Ingreso total de cada tienda:")
print(f"Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Tienda 4: ${ingreso_tienda4:,.2f}")
```

# 2. Ventas por categoría

```python
ingreso_tienda = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print("Ingreso total de cada tienda:")
print(f"Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Tienda 4: ${ingreso_tienda4:,.2f}")
```

# 3. Calificación promedio de la tienda


```python
print("Productos vendidos por categoría - Tienda 1:")
print(tienda.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False))

print("\nTienda 2:")
print(tienda2.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False))

print("\nTienda 3:")
print(tienda3.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False))

print("\nTienda 4:")
print(tienda4.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False))
```

# 4. Productos más y menos vendidos

```python
calif_tienda1 = tienda['Calificación'].mean()
calif_tienda2 = tienda2['Calificación'].mean()
calif_tienda3 = tienda3['Calificación'].mean()
calif_tienda4 = tienda4['Calificación'].mean()

print("Calificación promedio de cada tienda:")
print(f"Tienda 1: {calif_tienda1:.2f}")
print(f"Tienda 2: {calif_tienda2:.2f}")
print(f"Tienda 3: {calif_tienda3:.2f}")
print(f"Tienda 4: {calif_tienda4:.2f}")
```

# 5. Envío promedio por tienda

```python
ventas_prod1 = tienda.groupby('Producto')['Producto'].count().sort_values(ascending=False)
print("Tienda 1 - Producto más vendido:")
print(ventas_prod1.head(1))
print("Tienda 1 - Producto menos vendido:")
print(ventas_prod1.tail(1))

ventas_prod2 = tienda2.groupby('Producto')['Producto'].count().sort_values(ascending=False)
print("\nTienda 2 - Producto más vendido:")
print(ventas_prod2.head(1))
print("Tienda 2 - Producto menos vendido:")
print(ventas_prod2.tail(1))

ventas_prod3 = tienda3.groupby('Producto')['Producto'].count().sort_values(ascending=False)
print("\nTienda 3 - Producto más vendido:")
print(ventas_prod3.head(1))
print("Tienda 3 - Producto menos vendido:")
print(ventas_prod3.tail(1))

# Tiena 4
ventas_prod4 = tienda4.groupby('Producto')['Producto'].count().sort_values(ascending=False)
print("\nTienda 4 - Producto más vendido:")
print(ventas_prod4.head(1))
print("Tienda 4 - Producto menos vendido:")
print(ventas_prod4.tail(1))
```

```python
ingresos = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]
tiendas_nombres = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.figure(figsize=(8,5))
sns.barplot(x=tiendas_nombres, y=ingresos, palette="viridis")
plt.title("Ingreso total por tienda")
plt.ylabel("Ingreso total ($)")
plt.show()
```
